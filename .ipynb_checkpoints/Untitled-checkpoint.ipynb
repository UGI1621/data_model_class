```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data.csv")
```

```python
print(df)
```

```python
Total_Revenue = df.iloc[0][1:]
print(Total_Revenue)
```

```python
Target_Revenue = df.iloc[1][1:]
print(Target_Revenue)
```

```python
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],Total_Revenue,label='Total_Revenue')
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],Target_Revenue,label='target_Revenue')

plt.title('Monthly Revenue whit Target Revenue')
plt.xlabel('Month')
plt.ylabel('Total Rvenue')
plt.legend()

plt.show()
```

```python
Advertising_Costs = df.iloc[2][1:]
Social_Networking_Costs = df.iloc[3][1:]
print("----------Advertising_Costs----------")
print(Advertising_Costs)
print()
print("----------Social_Networking_Costs----------")
print(Social_Networking_Costs)
```

```python
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],Advertising_Costs,label='Advertising_Costs')
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],Social_Networking_Costs,label='Social_Networking_Costs')

plt.title('Advertising_Costs VS. Social_Networking_Costs')
plt.xlabel('Month')
plt.ylabel('Costs')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],Total_Revenue,label='Total_Revenue')
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],Target_Revenue,label='target_Revenue')

plt.title('Monthly Revenue whit Target Revenue')
plt.xlabel('Month')
plt.ylabel('Total Rvenue')
plt.legend()


plt.subplot(1,2,2)
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],Advertising_Costs,label='Advertising_Costs')
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],Social_Networking_Costs,label='Social_Networking_Costs')

plt.title('Advertising_Costs VS. Social_Networking_Costs')
plt.xlabel('Month')
plt.ylabel('Costs')
plt.legend()

plt.show()
```
